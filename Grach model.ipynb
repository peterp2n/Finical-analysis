```python
# Define the ticker symbol for the S&P 500
tickerSymbol = "^GSPC"

# Define the time period for which you want to download the data
startDate = '2018-01-01'
endDate = '2021-12-31'

# Download the data using the yfinance library
spx = yf.download(tickerSymbol, start=startDate, end=endDate)
spx
```

```python
daily_close_spx = spx[['Adj Close']]
daily_pct_spxchange = daily_close_spx / daily_close_spx.shift(1) - 1
daily_pct_spxchange.fillna(0, inplace=True) # in case error
x = daily_pct_spxchange # returns of spx
x = x.drop(x.head(1).index) * 100
print ('daily_close_spx' ,daily_close_spx)
```

```python
# Fit an ARCH(1) model
model_1 = arch_model(x, vol='ARCH', p=1)
result_1 = model_1.fit(update_freq = 2)
result_1.summary
```

```python
# Fit an GARCH(1,1) model
model_2 = arch_model(x, vol='GARCH', p=1, o=0, q=1)
result_2 = model_2.fit(update_freq = 2)
result_2.summary
```

```python
# Fit an GARCH(1,2) model
model_3 = arch_model(x, vol='GARCH', p=1, o=0, q=2)
result_3 = model_3.fit(update_freq = 2)
result_3.summary
```

```python
# Fit an GARCH(2,1) model
model_4 = arch_model(x, vol='GARCH', p=2, o=0, q=1)
result_4 = model_4.fit(update_freq = 2)
result_4.summary
```
